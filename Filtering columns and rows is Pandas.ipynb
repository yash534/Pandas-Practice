# Filtering coloumns and rows in pandas.

```python
import pandas as pd
```

```python
df = pd.read_excel(r"Y:\DataSets\emp-data.xlsx")
```

```python
df
```

```python
df[df['Age'] <= 30 ]
```

```python
specific_departments = ['Sales', 'Website']

df[df['Department'].isin(specific_departments)]
```

```python
df[df['Department'].str.contains('Finance')]
```

```python
df2 = df.set_index('Department')
```

```python
df2
```

```python
df2.filter(items= ['Name', 'Salary'])
```

```python
df2.filter(like = 'HR', axis = 0)
```

```python
df2.loc['Website']
```

```python
df2.iloc[7]
```

```python
df[df['Age'] <= 30].sort_values(by = [ 'Age','Salary'], ascending = False)
```
